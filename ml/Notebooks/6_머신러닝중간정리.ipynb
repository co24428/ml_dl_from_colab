# 개요

- Machine Learning
- 데이터 => 정보 => 지식 => 지혜
- 데이터로부터 지식/지혜(통창 etc)를 추출하는 작업
- 통계학, 컴퓨터 사이언스, 데이터마이닝, 인공지능 등이 다 섞여있다.
- 예측분석, 통계적 머신러닝 등등 칭하는 명칭도 다양

# 정의

- 명시적 프로그래밍 단계가 없이 컴퓨터가 스스로 학습을 하게 하는 능력을 갖추게 하는 연구분야이다.
- 예) 스팸필터
  - T : Task, 새로운 메일이 스팸인지 아닌지를 구분
  - E : Experience, 훈련 데이터 (시스템이 학습하는 샘플)
  - P : Perperence, 정확하게 스팸 메일을 걸러내는 비율 -> 정확도
  - 이 프로그램은 작업 T와 성능 측정 P에 대한경험 E로 학습한 것이다.

# 머신러닝 시스템 종류

- 학습하는동안의 감독형태, 정보량, 정보의 형태에 따라 나눠진다.

<hr>

- 지도학습 : supervised learning
  - 분류
    - 스팸 필터 (스팸이다, 아니다)
  - 회귀
    - 중고차 가격 예측(특성: 주행거리, 연식, 브랜드, 사고이력, ...), 이를 통한 수치 예측
  - 알고리즘
    - K-최근접 이웃, 선형회귀, 로지스틱회귀, 서포트벡터머신, 결정트리, 랜덤포레스트, XG부스트, 신경망

<hr>

- 비지도학습 : unsupervised learning
  - 훈련데이터에 레이블이 없다.
  - 예시
    - 블로그/사이트 방문자 그룹화
      - 알고리즘 스스로가 방문자 사이의 연결고리를 찾는다
      - 예) 방문자의 50%가 기혼자, 이들은 영화장르를 선호한다  
        전체 방문자의 80%는 21시~22시 사이에 많이 들어온다.  
        국가별 방문기록 등등 ...  
        => 군집화, 유형을 찾아내고 생성
      - 시각화
        - 고차원 데이터를 삽입하여 도식화
        - 2d/3d 등으로 표현 -> 알고리즘은 구조를 유지할려고 하는 경향을 가진다.
        - 패턴을 추출
      - 차원축소
        - 정보의 손실을 최소화 -> 데이터를 간소화
        - 여러 특성을 하나의 특성으로 합치는 과정
          - 예) 아이리스의 특성이 4개인데, 2개로 줄인다.
          - 특성이 많으면 적용할만한 사안으로 보인다.
          - 예) 주행거리, 연식 관련 데이터를 차원축소를 통해 => 마모도라는 특성으로 합친다.
        - 실행/학습 속도 상승 => 디스크/메모리상 연산상 점유율을 줄이고  
          생산성이 향상, 정확도 상승될 수도 있다.
      - 이상치 탐지
        - 부정 거래 탐지
        - 부정 대출 탐지
        - 제조 결함 감지 (품질 관리, 금융, IOT -> 시계열 -> RNN + LSTM)
      - 연관규칙 학습
        - 대량의 데이터에서 특성간의 흥미로운 관계를 찾아내는 것
        - 편의점/마트/수퍼마켓/온라인상점  
          => POS( VAN사(VISA, BC),  
          PG사(이니시스, LG),  
          서브PG사,  
          POS회사(한국정보통신) => 물품들 간의 연관규칙을 찾을 수 있다  
          => 오프라인이면 매대 배치, 온라인이면 마케팅 등등 적용

<hr>

- 준지도학습 : semi-supervised learning
  - 레이블이 일부만 존재
  - 레이블이 있는 데이터는 소수
  - 지도학습 + 비지도학습 연계 => 여러 모형을 섞어서 구현

<hr>

- 강화학습 : Reinforcement learning
  - 딥러닝 떄 표현
  - 에이전트 : 학습을 수행하는 시스템
  - 리워드(긍정적 보상), 패널티(부정적 보상)
  - 예) 딥마인드의 알파고 -> 알파제로
    - 수백/수천만/수억의 게임을 분석 => 승리에 대한 학습 => 대국 중에도 속해서 알고리즘을 발전시키면서 대전(이세돌 이벤트시, 알파고)
    - 알파제로 => 이미 성립된 그간의 전략을 이용하여, 학습은 오프라인방식(대국 중에는 따로 학습X) => 커제 압승


# 학습법

## 입력 데이터의 스트림(데이터의 연속적인 흐름)으로부터 점직적 학습이 가능한가?


- 1. 오프라인 학습 : 배치 학습
  - 새로운 데이터를 학습에 적용하기 위해서
    - 전체 데이터(이전데이터 + 신규데이터)를 사용하여 새로운 버전으로 다시 학습/훈련
    - 적용방법
      - 시스템 중단, 새로운 알고리즘으로 교체, 시스템 가동 => restart(재가동)
    - 주기
      - 24시간, 매주, 매월 => 사전에 교체에 대한 연습/시뮬레이션!! => 자동화고민
    - 고려사항
      - 시간이 많이 지나면!!
        - 데이터가 쌓여간다 -> 학습 비용이 상승(돈/시간) -> 언젠가는 불가능하게 될것이다  
          -> 어느 정도 데이터가 많이지면(사이트별로 다르다) -> 온라인 학습법으로 넘어가야 한다
      - 자원이 제한된 시스템인 경우
        - 스마트폰, 화성 탐사 로봇
        - 학습에 모든 리소스를 할당할수 없다. 자원의 분배가 필요
      - 시스템훈련비용
      - 제품에 적용하는 방식
      - 서비스 초기, 어느정도 데이터가 구출될때까지는 OK
      - 롤백기능도 준비(v1.0.1 -> v1.0.0 되돌릴수 있는가?)

- 2. 온라인 학습 : 세미배치학습, 점진적 학습법 : incremental learning
  - 데이터를 순차적으로 미니 batch(데이터량) 단위로 기존 알고리즘에 주입하여 시스템(알고리즘)을 훈련시킨다
  - 매단계 빠르고, 비용도 적게 든다
  - 데이터가 도착하는 대로(한개 ~ n개) 즉시 적용 => 게임, 주식, 자율주행 등등 빠른 변화에 대응하는 7일후
  - 시스템에 적합
  - 자원이 한정적이다 => 학습이 끝나면 데이터(로그처리)는 버린다
    - 롤백 가능하게 구성, 완전 삭제까지는 7일후?
  - 데이터는 외부 디스크상에서 읽고 => 주입 => 로드 => 학습 => 버린다
  - 단점 : 나쁜 데이터가 주입되면 => 시스템 성능이 저하된다!!
    - 이상탐지, 성능평가등 데이터의 대한 평가가 수행되고 나서 주입
    - 롤백기능이 강화

# 방식
- 목표 : 새로운 샘플(한번도 접하지 못한 데이터)이 잘 작동하게 하는 모델을 구축
- 사례 기반 학습
  - 목표를 단순하게 기억하는 방식
  - 스펨 메일-> 스펨으로 분류되는 모든 메일을 기억한다 -> 동일 유형 분류해 낸다
  - 사례를 많이 기억하여 학습을 수행
- 모델 기반 학습
  - 전체데이터의 샘플로 모델을 만들어서 예측을 사용
